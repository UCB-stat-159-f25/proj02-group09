### Read Data

```python
# imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn-v0_8-dark') 
```

```python
sou = pd.read_csv("data/SOTU.csv")
sou
```

#### Number of Speeches per President

```python
speeches_per_president = sou['President'].value_counts().reindex(sou['President'].unique())
```

```python
plt.figure(figsize=(14, 5))
speeches_per_president.plot(kind='bar', color='C0')
plt.title('Number of Speeches per President')
plt.xlabel('President')
plt.ylabel('Count')
# From the graph we can see that most Presidents tend to have 4 speeches, 
# with George Washington having the most at 12
# and Zachary Taylor having the least at 1.
```

#### Number of Speeches per Year

```python
speeches_per_year = sou['Year'].value_counts().sort_index()
plt.figure(figsize=(7, 5))
speeches_per_year.plot(kind='line', color='C0')
plt.title('Number of State of the Union Speechs per Year')
plt.xlabel('Year')
plt.ylabel('Count')
# It is interesting to see that the plot started at the count of 4 
# and quickly dropped to 2 and remained at 1 for 150 years before jumping up to 2 again.
```

#### Word Count Distribution 

```python
plt.figure(figsize=(7, 5))
sns.histplot(sou['Word Count'], color='C0')
plt.title('Distribution of State of the Union Speech Word Counts')
plt.xlabel('Word Count')
plt.ylabel('Count')
# From the plot we can see that the distribution is right skewed, 
# which means that most of the speeches are relatively short under 10,000 words.
```

#### Word Count Distribution over Year

```python
plt.figure(figsize=(7, 5))
sns.scatterplot(x=sou['Word Count'], y=sou['Year'], color='C0')
sns.rugplot(x=sou['Word Count'], color='C0')
sns.rugplot(y=sou['Year'], color='C0')
plt.margins(y=0.06)
plt.title('Speech Year Versus Word Count')
plt.xlabel('Word Count')
plt.ylabel('Year')
# From the plot we can see in recent years, the speech word count has gone down,
# we can also see a single data point on the far right, being an outlier.
```

#### Word Count Distribution per President

```python
mean_word_count_sorted = sou.groupby('President')['Word Count'].mean().reindex(sou['President'].unique())
plt.figure(figsize=(14, 5))
mean_word_count_sorted.plot(kind='bar', color='C0')
plt.title('Average State of the Union Word Count per President')
plt.xlabel('President')
plt.ylabel('Average Word Count')
# It is interesting to see that most recent presidents tend to have lower average word counts
# while more historical presidents have higher word counts.
```
